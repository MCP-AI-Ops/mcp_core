```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, regularizers, callbacks
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler, RobustScaler
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# GPU 최적화 설정
print("TensorFlow 버전:", tf.__version__)
print("GPU 사용 가능:", tf.config.list_physical_devices('GPU'))

# 재현 가능성을 위한 시드 설정
def set_random_seeds(seed=42):
    np.random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

set_random_seeds(42)

file_path = "/content/drive/MyDrive/MCPproject/lstm_ready_cluster_data.csv"
print(f"파일경로 : {file_path}")
print(os.path.exists(file_path))

class CompleteMCPPredictor:
    """
    Google Cluster 데이터의 극값과 long-tail 분포 특성을 고려한 설계
    """

    def __init__(self,
                 sequence_length=24,
                 target_col='total_events',
                 test_size=0.2,
                 val_size=0.1,
                 use_log_transform=True,
                 handle_outliers=True):

        self.seq_len = sequence_length
        self.target_col = target_col
        self.test_size = test_size
        self.val_size = val_size
        self.use_log_transform = use_log_transform
        self.handle_outliers = handle_outliers

        # 스케일러들
        self.feature_scaler = RobustScaler()  # 극값에 강건한 스케일러
        self.target_scaler = RobustScaler()

        # 모델과 데이터
        self.model = None
        self.history = None
        self.feature_names = None

        # 데이터 저장
        self.X_train = self.X_val = self.X_test = None
        self.y_train = self.y_val = self.y_test = None
        self.df = None

    def load_and_prepare_data(self, csv_path='/content/drive/MyDrive/MCPproject/lstm_ready_cluster_data.csv'):
        """
        데이터 로드 및 전처리
        """
        if csv_path == 'lstm_ready_cluster_data.csv':
            csv_path = '/content/drive/MyDrive/MCPproject/lstm_ready_cluster_data.csv'
            print(f"파일명만 전달되어 전체 경로로 변환: {csv_path}")

        # 파일이 없으면 샘플 데이터 생성
        if not os.path.exists(csv_path):
            print(f"파일 {csv_path}를 찾을 수 없습니다. 샘플 데이터를 생성합니다.")
            self.df = self._create_sample_data()
        else:
            self.df = pd.read_csv(csv_path)
            print(f"데이터 로드 완료: {self.df.shape}")

        # 기본 정리
        self.df = self.df.sort_values('hour_offset').reset_index(drop=True)

        # 타겟 변수 전처리
        self._preprocess_target()

        # 특성 선택
        self._select_features()

        # 이상치 처리
        if self.handle_outliers:
            self._handle_outliers()

        # 결측값 처리
        self._handle_missing_values()

        print(f"전처리 완료: {self.df.shape}")
        print(f"타겟 통계 - 평균: {self.df[self.target_col].mean():.2f}, "
              f"분산: {self.df[self.target_col].var():.2f}")

        return self.df

    def _create_sample_data(self, n_points=1000):
        """
        샘플 데이터 생성 (파일이 없을 때 사용)
        """
        print("샘플 데이터 생성 중...")

        np.random.seed(42)
        hours = np.arange(n_points)

        # 복합적인 시계열 패턴 생성
        daily_pattern = 30 * np.sin(2 * np.pi * hours / 24) + 20
        weekly_pattern = 15 * np.sin(2 * np.pi * hours / (24*7))
        trend = 0.02 * hours
        noise = np.random.normal(0, 10, n_points)

        # 극값 추가 (실제 클라우드 패턴 모방)
        spike_indices = np.random.choice(n_points, size=n_points//50, replace=False)
        spikes = np.zeros(n_points)
        spikes[spike_indices] = np.random.exponential(200, len(spike_indices))

        total_events = np.maximum(daily_pattern + weekly_pattern + trend + noise + spikes, 1)

        # 관련 특성들 생성
        unique_machines = total_events * 0.1 + np.random.normal(0, 5, n_points)
        unique_machines = np.maximum(unique_machines, 1)

        avg_cpu = 0.3 + 0.4 * (total_events / total_events.max()) + np.random.normal(0, 0.1, n_points)
        avg_cpu = np.clip(avg_cpu, 0.1, 1.0)

        avg_memory = 0.2 + 0.5 * (total_events / total_events.max()) + np.random.normal(0, 0.08, n_points)
        avg_memory = np.clip(avg_memory, 0.1, 1.0)

        # 시간 특성
        hour_of_day = hours % 24
        day_of_week = (hours // 24) % 7

        df = pd.DataFrame({
            'hour_offset': hours,
            'total_events': total_events,
            'unique_machines': unique_machines,
            'avg_cpu': avg_cpu,
            'avg_memory': avg_memory,
            'hour_of_day': hour_of_day,
            'day_of_week': day_of_week,
        })

        # 추가 특성 생성
        for lag in [1, 2, 3, 6, 12, 24]:
            df[f'lag{lag}_events'] = df['total_events'].shift(lag).fillna(df['total_events'].mean())

        for window in [3, 6, 12, 24]:
            df[f'ma{window}_events'] = df['total_events'].rolling(window, min_periods=1).mean()

        # 순환 특성
        df['hour_sin'] = np.sin(2 * np.pi * df['hour_of_day'] / 24)
        df['hour_cos'] = np.cos(2 * np.pi * df['hour_of_day'] / 24)
        df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
        df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)

        print(f"샘플 데이터 생성 완료: {df.shape}")
        return df

    def _preprocess_target(self):
        """
        타겟 변수 전처리 (극값 고려)
        """
        original_target = self.df[self.target_col].copy()

        if self.use_log_transform:
            # 로그 변환으로 극값의 영향 감소
            self.df[f'{self.target_col}_log'] = np.log1p(self.df[self.target_col])
            self.target_col = f'{self.target_col}_log'
            print(f"로그 변환 적용: {original_target.mean():.1f} -> {self.df[self.target_col].mean():.3f}")

        # 분위수 기반 극값 제한 (옵션)
        if self.handle_outliers:
            q95 = self.df[self.target_col].quantile(0.95)
            q5 = self.df[self.target_col].quantile(0.05)

            outliers_high = (self.df[self.target_col] > q95).sum()
            outliers_low = (self.df[self.target_col] < q5).sum()

            if outliers_high > 0 or outliers_low > 0:
                self.df[self.target_col] = self.df[self.target_col].clip(lower=q5, upper=q95)
                print(f"극값 제한: 상위 {outliers_high}개, 하위 {outliers_low}개 조정")

    def _select_features(self):
        """
        예측에 사용할 특성 선택
        """
        # 제외할 컬럼들
        exclude_cols = {'hour_offset', 'total_events', 'total_events_log'}

        # 수치형 특성들만 선택
        numeric_cols = self.df.select_dtypes(include=[np.number]).columns
        feature_cols = [col for col in numeric_cols if col not in exclude_cols]

        # 분산이 0인 특성 제거
        zero_var_cols = []
        for col in feature_cols:
            if self.df[col].std() == 0:
                zero_var_cols.append(col)

        if zero_var_cols:
            print(f"분산 0 특성 제거: {zero_var_cols}")
            feature_cols = [col for col in feature_cols if col not in zero_var_cols]

        self.feature_names = feature_cols
        print(f"선택된 특성 수: {len(self.feature_names)}")

    def _handle_outliers(self):
        """
        특성별 이상치 처리
        """
        outlier_count = 0
        for col in self.feature_names:
            Q1 = self.df[col].quantile(0.25)
            Q3 = self.df[col].quantile(0.75)
            IQR = Q3 - Q1

            if IQR > 0:
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR

                outliers = ((self.df[col] < lower_bound) | (self.df[col] > upper_bound)).sum()
                if outliers > 0:
                    self.df[col] = self.df[col].clip(lower=lower_bound, upper=upper_bound)
                    outlier_count += outliers

        if outlier_count > 0:
            print(f"특성 이상치 처리: {outlier_count}개 값 조정")

    def _handle_missing_values(self):
        """
        결측값 처리
        """
        # 시계열 순서 보장
        self.df = self.df.sort_values('hour_offset')

        # LAG 특성들은 forward fill
        lag_cols = [col for col in self.df.columns if 'lag' in col]
        for col in lag_cols:
            self.df[col] = self.df[col].fillna(method='ffill')

        # 이동평균 특성들은 backward fill
        ma_cols = [col for col in self.df.columns if 'ma' in col]
        for col in ma_cols:
            self.df[col] = self.df[col].fillna(method='bfill')

        # 나머지는 선형 보간 후 평균값
        self.df = self.df.interpolate().fillna(self.df.mean())

        # 무한값 처리
        self.df = self.df.replace([np.inf, -np.inf], np.nan).fillna(0)

    def create_sequences(self):
        """
        시계열 시퀀스 생성
        """
        print("=== 시퀀스 데이터 생성 ===")

        # 시간 기반 분할
        n = len(self.df)
        train_end = int(n * (1 - self.test_size - self.val_size))
        val_end = int(n * (1 - self.test_size))

        df_train = self.df.iloc[:train_end].copy()
        df_val = self.df.iloc[train_end:val_end].copy()
        df_test = self.df.iloc[val_end:].copy()

        print(f"데이터 분할: Train({len(df_train)}), Val({len(df_val)}), Test({len(df_test)})")

        # 특성 정규화 (훈련 데이터 기준)
        train_features = df_train[self.feature_names].values
        self.feature_scaler.fit(train_features)

        # 타겟 정규화 (훈련 데이터 기준)
        train_targets = df_train[self.target_col].values.reshape(-1, 1)
        self.target_scaler.fit(train_targets)

        # 각 분할에 대해 시퀀스 생성
        self.X_train, self.y_train = self._build_sequences(df_train)
        self.X_val, self.y_val = self._build_sequences(df_val)
        self.X_test, self.y_test = self._build_sequences(df_test)

        print(f"시퀀스 생성 완료:")
        print(f"  Train: X{self.X_train.shape}, y{self.y_train.shape}")
        print(f"  Val:   X{self.X_val.shape}, y{self.y_val.shape}")
        print(f"  Test:  X{self.X_test.shape}, y{self.y_test.shape}")

        # 시퀀스가 부족한 경우 경고
        min_sequences = 100
        if self.X_train.shape[0] < min_sequences:
            print(f"경고: 훈련 시퀀스가 {self.X_train.shape[0]}개로 부족합니다. 최소 {min_sequences}개 권장")

    def _build_sequences(self, df):
        """
        단일 데이터프레임에서 시퀀스 생성
        """
        # 특성 정규화
        features = self.feature_scaler.transform(df[self.feature_names].values)

        # 타겟 정규화
        targets = self.target_scaler.transform(df[self.target_col].values.reshape(-1, 1)).flatten()

        # 슬라이딩 윈도우로 시퀀스 생성
        X, y = [], []
        for i in range(self.seq_len, len(df)):
            X.append(features[i-self.seq_len:i])
            y.append(targets[i])

        return np.array(X, dtype=np.float32), np.array(y, dtype=np.float32)

    def build_model(self,
                   lstm_units=[64, 32],
                   dense_units=[16, 8],
                   dropout_rate=0.3,
                   l2_reg=1e-4,
                   learning_rate=0.001):
        """
        개선된 LSTM 모델 구축
        """
        print("=== LSTM 모델 구축 ===")

        n_features = len(self.feature_names)

        model = tf.keras.Sequential([
            layers.Input(shape=(self.seq_len, n_features)),

            # 첫 번째 LSTM 레이어
            layers.LSTM(lstm_units[0],
                       return_sequences=True,
                       dropout=dropout_rate,
                       recurrent_dropout=dropout_rate/2,
                       kernel_regularizer=regularizers.l2(l2_reg)),
            layers.BatchNormalization(),

            # 두 번째 LSTM 레이어
            layers.LSTM(lstm_units[1],
                       return_sequences=False,
                       dropout=dropout_rate,
                       recurrent_dropout=dropout_rate/2,
                       kernel_regularizer=regularizers.l2(l2_reg)),
            layers.BatchNormalization(),

            # Dense 레이어들
            layers.Dense(dense_units[0], activation='relu',
                        kernel_regularizer=regularizers.l2(l2_reg)),
            layers.Dropout(dropout_rate),
            layers.BatchNormalization(),

            layers.Dense(dense_units[1], activation='relu',
                        kernel_regularizer=regularizers.l2(l2_reg)),
            layers.Dropout(dropout_rate/2),

            layers.Dense(1)  # 선형 출력
        ])

        # 옵티마이저 설정 (고정 학습률 사용)
        optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

        # Huber loss 사용 (극값에 강건)
        model.compile(
            optimizer=optimizer,
            loss='huber',  # 극값에 덜 민감
            metrics=['mae', 'mse']
        )

        self.model = model

        print("모델 구조:")
        model.summary()

        return model

    def train_model(self,
                   epochs=150,
                   batch_size=32,
                   patience=25,
                   verbose=1):
        print("=== 모델 훈련 시작 ===")

        # 효과적인 배치 크기 계산
        effective_batch_size = min(batch_size, max(8, self.X_train.shape[0] // 10))

        # 콜백 설정
        callbacks_list = [
            callbacks.EarlyStopping(
                monitor='val_loss',
                patience=patience,
                restore_best_weights=True,
                verbose=1
            ),
            callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                patience=patience//2,
                factor=0.5,
                min_lr=1e-7,
                verbose=1
            ),
            callbacks.ModelCheckpoint(
                'best_mcp_lstm_model.h5',
                monitor='val_loss',
                save_best_only=True,
                verbose=1
            )
        ]

        # 검증 데이터 설정
        validation_data = None
        if self.X_val.shape[0] > 0:
            validation_data = (self.X_val, self.y_val)
        else:
            print("검증 데이터가 부족하여 훈련 데이터의 20%를 사용")
            validation_split = 0.2

        # 훈련 실행
        history = self.model.fit(
            self.X_train, self.y_train,
            validation_data=validation_data,
            validation_split=0.2 if validation_data is None else 0.0,
            epochs=epochs,
            batch_size=effective_batch_size,
            callbacks=callbacks_list,
            verbose=verbose,
            shuffle=True
        )

        self.history = history
        print("모델 훈련 완료")

        return history

    def evaluate_model(self):
        """
        종합적인 모델 평가
        """
        print("=== 모델 평가 ===")

        results = {}

        # 각 데이터셋에 대해 예측 및 평가
        datasets = [
            ('Train', self.X_train, self.y_train),
            ('Validation', self.X_val, self.y_val),
            ('Test', self.X_test, self.y_test)
        ]

        for name, X, y_true in datasets:
            if X.shape[0] == 0:
                continue

            # 예측
            y_pred_scaled = self.model.predict(X, verbose=0).flatten()

            # 원래 스케일로 복원
            y_pred = self.target_scaler.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
            y_actual = self.target_scaler.inverse_transform(y_true.reshape(-1, 1)).flatten()

            # 로그 변환 역변환
            if self.use_log_transform:
                y_pred = np.expm1(y_pred)
                y_actual = np.expm1(y_actual)

            # 메트릭 계산
            mae = mean_absolute_error(y_actual, y_pred)
            rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
            r2 = r2_score(y_actual, y_pred)

            # MAPE 계산 (0으로 나누기 방지)
            mape = np.mean(np.abs((y_actual - y_pred) / np.maximum(y_actual, 1))) * 100

            results[name.lower()] = {
                'mae': mae,
                'rmse': rmse,
                'r2': r2,
                'mape': mape,
                'y_true': y_actual,
                'y_pred': y_pred
            }

            print(f"{name:>10}: MAE={mae:.2f}, RMSE={rmse:.2f}, R²={r2:.4f}, MAPE={mape:.1f}%")

        return results

    def plot_comprehensive_results(self, results):
        """
        종합적인 결과 시각화
        """
        fig = plt.figure(figsize=(20, 16))

        # 1. 훈련 히스토리
        if self.history:
            ax1 = plt.subplot(3, 4, 1)
            plt.plot(self.history.history['loss'], label='Train Loss', linewidth=2)
            if 'val_loss' in self.history.history:
                plt.plot(self.history.history['val_loss'], label='Val Loss', linewidth=2)
            plt.title('Training History - Loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend()
            plt.grid(True, alpha=0.3)

            ax2 = plt.subplot(3, 4, 2)
            plt.plot(self.history.history['mae'], label='Train MAE', linewidth=2)
            if 'val_mae' in self.history.history:
                plt.plot(self.history.history['val_mae'], label='Val MAE', linewidth=2)
            plt.title('Training History - MAE')
            plt.xlabel('Epoch')
            plt.ylabel('MAE')
            plt.legend()
            plt.grid(True, alpha=0.3)

        # 2. 예측 vs 실제 (테스트 데이터)
        if 'test' in results:
            y_true = results['test']['y_true']
            y_pred = results['test']['y_pred']
            r2 = results['test']['r2']

            ax3 = plt.subplot(3, 4, 3)
            plt.scatter(y_true, y_pred, alpha=0.6, s=20)
            min_val, max_val = min(np.min(y_true), np.min(y_pred)), max(np.max(y_true), np.max(y_pred))
            plt.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction')
            plt.xlabel('Actual Values')
            plt.ylabel('Predicted Values')
            plt.title(f'Test: Actual vs Predicted (R²={r2:.3f})')
            plt.legend()
            plt.grid(True, alpha=0.3)

            # 3. 잔차 분석
            ax4 = plt.subplot(3, 4, 4)
            residuals = y_true - y_pred
            plt.scatter(y_pred, residuals, alpha=0.6, s=20)
            plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
            plt.xlabel('Predicted Values')
            plt.ylabel('Residuals')
            plt.title('Residual Analysis')
            plt.grid(True, alpha=0.3)

        # 4. 시계열 예측 비교
        if 'test' in results:
            y_true = results['test']['y_true']
            y_pred = results['test']['y_pred']

            ax5 = plt.subplot(3, 4, 5)
            sample_size = min(100, len(y_true))
            idx = slice(0, sample_size)

            plt.plot(y_true[idx], label='Actual', linewidth=2, alpha=0.8)
            plt.plot(y_pred[idx], label='Predicted', linewidth=2, alpha=0.8)
            plt.title(f'Time Series Prediction (First {sample_size} points)')
            plt.xlabel('Time Steps')
            plt.ylabel('Values')
            plt.legend()
            plt.grid(True, alpha=0.3)

        # 5. 오차 분포
        if 'test' in results:
            ax6 = plt.subplot(3, 4, 6)
            residuals = results['test']['y_true'] - results['test']['y_pred']
            plt.hist(residuals, bins=30, alpha=0.7, edgecolor='black', density=True)
            plt.axvline(0, color='red', linestyle='--', linewidth=2, label='Zero Error')
            plt.xlabel('Residuals')
            plt.ylabel('Density')
            plt.title('Error Distribution')
            plt.legend()
            plt.grid(True, alpha=0.3)

        # 6-8. 각 데이터셋별 성능 비교
        metrics_data = []
        for name, data in results.items():
            metrics_data.append([name.title(), data['mae'], data['rmse'], data['r2'], data['mape']])

        if metrics_data:
            ax7 = plt.subplot(3, 4, 7)
            datasets = [row[0] for row in metrics_data]
            maes = [row[1] for row in metrics_data]
            plt.bar(datasets, maes, alpha=0.7)
            plt.title('MAE by Dataset')
            plt.ylabel('MAE')
            plt.xticks(rotation=45)

            ax8 = plt.subplot(3, 4, 8)
            r2s = [row[3] for row in metrics_data]
            colors = ['green' if r2 > 0.5 else 'orange' if r2 > 0 else 'red' for r2 in r2s]
            plt.bar(datasets, r2s, alpha=0.7, color=colors)
            plt.title('R² Score by Dataset')
            plt.ylabel('R² Score')
            plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)
            plt.xticks(rotation=45)

        # 9. 예측 정확도 구간별 분석
        if 'test' in results:
            ax9 = plt.subplot(3, 4, 9)
            y_true = results['test']['y_true']
            y_pred = results['test']['y_pred']

            # 구간별 정확도 분석
            bins = np.percentile(y_true, [0, 25, 50, 75, 100])
            bin_labels = ['Q1', 'Q2', 'Q3', 'Q4']
            bin_accuracies = []

            for i in range(len(bins)-1):
                mask = (y_true >= bins[i]) & (y_true < bins[i+1])
                if i == len(bins)-2:  # 마지막 구간은 포함
                    mask = (y_true >= bins[i]) & (y_true <= bins[i+1])

                if mask.sum() > 0:
                    bin_r2 = r2_score(y_true[mask], y_pred[mask])
                    bin_accuracies.append(bin_r2)
                else:
                    bin_accuracies.append(0)

            plt.bar(bin_labels, bin_accuracies, alpha=0.7)
            plt.title('Accuracy by Value Range')
            plt.ylabel('R² Score')
            plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)

        # 10. 학습 진행 상황 (Loss vs Val Loss)
        if self.history and 'val_loss' in self.history.history:
            ax10 = plt.subplot(3, 4, 10)
            train_loss = self.history.history['loss']
            val_loss = self.history.history['val_loss']

            plt.plot(train_loss, label='Training Loss')
            plt.plot(val_loss, label='Validation Loss')

            # 과적합 여부 시각화
            min_loss_epoch = np.argmin(val_loss)
            min_val_loss = val_loss[min_loss_epoch]

            plt.axvline(x=min_loss_epoch, color='green', linestyle=':', alpha=0.7)
            plt.scatter([min_loss_epoch], [min_val_loss], color='green', s=100, zorder=5)

            plt.title('Training vs Validation Loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend()
            plt.grid(True, alpha=0.3)

        # 11. 성능 요약 테이블
        ax11 = plt.subplot(3, 4, 11)
        ax11.axis('off')

        if metrics_data:
            table_data = []
            table_data.append(['Dataset', 'MAE', 'RMSE', 'R²', 'MAPE(%)'])
            for row in metrics_data:
                formatted_row = [
                    row[0],
                    f'{row[1]:.2f}',
                    f'{row[2]:.2f}',
                    f'{row[3]:.3f}',
                    f'{row[4]:.1f}'
                ]
                table_data.append(formatted_row)

            table = ax11.table(cellText=table_data[1:], colLabels=table_data[0],
                              cellLoc='center', loc='center',
                              bbox=[0, 0, 1, 1])
            table.auto_set_font_size(False)
            table.set_fontsize(9)
            table.scale(1.2, 1.5)
            ax11.set_title('Performance Summary', pad=20)

        # 12. 특성 중요도 근사 (가중치 기반)
        ax12 = plt.subplot(3, 4, 12)
        if hasattr(self.model, 'layers') and len(self.model.layers) > 0:
            # 첫 번째 LSTM 레이어의 가중치로 특성 중요도 근사
            try:
                first_layer_weights = self.model.layers[0].get_weights()[0]  # input_to_hidden weights
                feature_importance = np.mean(np.abs(first_layer_weights), axis=1)

                # 상위 10개 특성만 표시
                top_n = min(10, len(feature_importance))
                top_indices = np.argsort(feature_importance)[-top_n:]
                top_features = [self.feature_names[i][:15] for i in top_indices]  # 이름 축약
                top_scores = feature_importance[top_indices]

                plt.barh(range(top_n), top_scores)
                plt.yticks(range(top_n), top_features)
                plt.xlabel('Approximate Importance')
                plt.title('Top Feature Importance')
                plt.grid(True, alpha=0.3)
            except:
                ax12.text(0.5, 0.5, 'Feature importance\nnot available',
                         ha='center', va='center', transform=ax12.transAxes)
                ax12.set_title('Feature Importance')

        plt.tight_layout()
        plt.savefig('complete_lstm_results.png', dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()

    def predict_future(self, steps=24):
        """
        미래 예측 (다중 스텝)
        """
        if self.X_test.shape[0] == 0:
            print("테스트 데이터가 없어 미래 예측을 수행할 수 없습니다.")
            return None

        print(f"미래 {steps}시간 예측 중...")

        # 마지막 시퀀스를 시작점으로 사용
        last_sequence = self.X_test[-1:].copy()
        predictions = []

        for step in range(steps):
            # 현재 시퀀스로 다음 값 예측
            pred_scaled = self.model.predict(last_sequence, verbose=0)[0, 0]

            # 원래 스케일로 복원
            pred_original = self.target_scaler.inverse_transform([[pred_scaled]])[0, 0]
            if self.use_log_transform:
                pred_original = np.expm1(pred_original)

            predictions.append(pred_original)

            # 시퀀스 업데이트 (예측값을 포함하여 다음 예측에 사용)
            new_features = last_sequence[0, -1:].copy()  # 마지막 타임스텝의 특성 복사
            new_features = np.concatenate([new_features, [[pred_scaled]]], axis=1)  # 예측값 추가 (임시)

            # 시퀀스를 한 스텝 앞으로 이동
            last_sequence = np.concatenate([last_sequence[:, 1:], new_features[:, :self.X_test.shape[2]].reshape(1, 1, -1)], axis=1)

        return np.array(predictions)

    def save_model_and_scaler(self, model_path='complete_mcp_lstm.h5'):
        """
        모델과 스케일러 저장
        """
        # 모델 저장
        self.model.save(model_path)

        # 스케일러와 메타데이터 저장
        import pickle

        metadata = {
            'feature_names': self.feature_names,
            'target_col': self.target_col,
            'sequence_length': self.seq_len,
            'use_log_transform': self.use_log_transform,
            'feature_scaler': self.feature_scaler,
            'target_scaler': self.target_scaler,
            'model_path': model_path
        }

        with open('mcp_model_metadata.pkl', 'wb') as f:
            pickle.dump(metadata, f)

        print(f"모델 저장: {model_path}")
        print("메타데이터 저장: mcp_model_metadata.pkl")

    def load_model_and_scaler(self, model_path='complete_mcp_lstm.h5', metadata_path='mcp_model_metadata.pkl'):
        """
        저장된 모델과 스케일러 로드
        """
        import pickle

        # 모델 로드
        self.model = tf.keras.models.load_model(model_path)

        # 메타데이터 로드
        with open(metadata_path, 'rb') as f:
            metadata = pickle.load(f)

        self.feature_names = metadata['feature_names']
        self.target_col = metadata['target_col']
        self.seq_len = metadata['sequence_length']
        self.use_log_transform = metadata['use_log_transform']
        self.feature_scaler = metadata['feature_scaler']
        self.target_scaler = metadata['target_scaler']

        print(f"모델 로드 완료: {model_path}")
        print(f"메타데이터 로드 완료: {metadata_path}")

# 메인 실행 함수
def main():
    """
    LSTM 파이프라인 실행
    """

    try:
        # 1. 모델 초기화
        predictor = CompleteMCPPredictor(
            sequence_length=24,
            target_col='total_events',
            test_size=0.2,
            val_size=0.15,
            use_log_transform=True,
            handle_outliers=True
        )

        # 2. 데이터 로드 및 전처리
        print("\n 데이터 로드 및 전처리")
        df = predictor.load_and_prepare_data('lstm_ready_cluster_data.csv')

        # 3. 시퀀스 생성
        print("\n 시퀀스 데이터 생성")
        predictor.create_sequences()

        # 4. 모델 구축
        print("\n LSTM 모델 구축")
        predictor.build_model(
            lstm_units=[64, 32],
            dense_units=[16, 8],
            dropout_rate=0.3,
            l2_reg=1e-4,
            learning_rate=0.001
        )

        # 5. 모델 훈련
        print("\n모델 훈련 시작...")
        history = predictor.train_model(
            epochs=100,
            batch_size=32,
            patience=20,
            verbose=1
        )

        # 6. 모델 평가
        print("\n 모델 평가")
        results = predictor.evaluate_model()

        # 7. 결과 시각화
        print("\n 결과 시각화")
        predictor.plot_comprehensive_results(results)

        # 8. 미래 예측 데모
        print("\n 미래 예측 데모")
        future_pred = predictor.predict_future(steps=24)
        if future_pred is not None:
            print(f"다음 24시간 예측 범위: {future_pred.min():.1f} ~ {future_pred.max():.1f}")

            # 미래 예측 시각화
            plt.figure(figsize=(12, 6))
            plt.plot(range(len(future_pred)), future_pred, 'o-', linewidth=2, markersize=4)
            plt.title('Future 24-Hour Prediction')
            plt.xlabel('Hours Ahead')
            plt.ylabel('Predicted Events')
            plt.grid(True, alpha=0.3)
            plt.tight_layout()
            plt.savefig('future_prediction.png', dpi=200, bbox_inches='tight')
            plt.show()

        # 9. 모델 저장
        print("\n 모델 저장")
        predictor.save_model_and_scaler()

        # 10. 최종 결과 요약
        print("\n" + "=" * 80)
        print(" 훈련 완료, 결과 요약:")
        print("=" * 80)

        for dataset, metrics in results.items():
            print(f"{dataset.upper():>12}: R²={metrics['r2']:.4f}, "
                  f"MAE={metrics['mae']:.2f}, MAPE={metrics['mape']:.1f}%")

        # R² 음수 확인
        test_r2 = results.get('test', {}).get('r2', -999)
        if test_r2 > 0.5:
            print(" R² > 0.5")
        elif test_r2 > 0:
            print(" 양수 R²달성, 성능 확보")
        else:
            print(" R² 음수. 모델 튜닝 필요")

        print("\n 생성된 파일:")
        print("  - complete_lstm_results.png: 종합 분석 결과")
        print("  - future_prediction.png: 미래 예측 결과")
        print("  - complete_mcp_lstm.h5: 훈련된 모델")
        print("  - mcp_model_metadata.pkl: 모델 메타데이터")
        print("  - best_mcp_lstm_model.h5: 최고 성능 모델")

        return predictor, results

    except Exception as e:
        print(f"\n 오류 발생: {str(e)}")
        import traceback
        traceback.print_exc()
        return None, None

# 코랩에서 실행
if __name__ == "__main__":
    predictor, results = main()

    if predictor is not None:
        print("\n 성공적으로 완료되었습니다!")
        print("이제 MCP 프로젝트에서 실시간 리소스 예측이 가능합니다.")
    else:
        print("\n 실행 중 문제가 발생했습니다. 로그를 확인하세요.")
```

```python
# 기존CompleteMCPPredictor 클래스에 24시간 예측 및 이상 탐지 기능 추가

# 필요한 추가 라이브러리
from sklearn.ensemble import IsolationForest
from datetime import datetime, timedelta
import json

class EnhancedLSTMPredictor(CompleteMCPPredictor):
    """
    기존 CompleteMCPPredictor에 24시간 예측과 이상 탐지 기능을 추가한 확장 클래스
    """

    def __init__(self,
                 sequence_length=24,
                 target_col='total_events',
                 test_size=0.2,
                 val_size=0.1,
                 use_log_transform=True,
                 handle_outliers=True,
                 prediction_horizon=24):  # 새로 추가된 파라미터

        # 부모 클래스 초기화
        super().__init__(sequence_length, target_col, test_size, val_size,
                        use_log_transform, handle_outliers)

        # 새로운 속성들
        self.prediction_horizon = prediction_horizon  # 예측할 시간 범위
        self.anomaly_detector = IsolationForest(contamination=0.1, random_state=42)
        self.prediction_errors = []
        self.context_weights = {
            'peak': 1.2,      # 피크시간 가중치
            'normal': 1.0,    # 정상시간
            'low': 0.8,       # 저사용시간
            'weekend': 0.7,   # 주말
            'holiday': 0.6    # 공휴일
        }

    def build_model(self,
                   lstm_units=[64, 32],
                   dense_units=[32, 16],
                   dropout_rate=0.3,
                   l2_reg=1e-4,
                   learning_rate=0.001):
        """
        24시간 예측을 위한 모델 구조 (기존 모델 확장)
        """
        print("=== 24시간 예측 LSTM 모델 구축 ===")

        n_features = len(self.feature_names)

        model = tf.keras.Sequential([
            layers.Input(shape=(self.seq_len, n_features)),

            # 첫 번째 LSTM 레이어
            layers.LSTM(lstm_units[0],
                       return_sequences=True,
                       dropout=dropout_rate,
                       recurrent_dropout=dropout_rate/2,
                       kernel_regularizer=regularizers.l2(l2_reg)),
            layers.BatchNormalization(),

            # 두 번째 LSTM 레이어
            layers.LSTM(lstm_units[1],
                       return_sequences=False,
                       dropout=dropout_rate,
                       recurrent_dropout=dropout_rate/2,
                       kernel_regularizer=regularizers.l2(l2_reg)),
            layers.BatchNormalization(),

            # Dense 레이어들
            layers.Dense(dense_units[0], activation='relu',
                        kernel_regularizer=regularizers.l2(l2_reg)),
            layers.Dropout(dropout_rate),
            layers.BatchNormalization(),

            layers.Dense(dense_units[1], activation='relu',
                        kernel_regularizer=regularizers.l2(l2_reg)),
            layers.Dropout(dropout_rate/2),

            # 출력 레이어 - 24시간 예측을 위해 변경
            layers.Dense(self.prediction_horizon)  # 24개 시점 동시 예측
        ])

        optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

        model.compile(
            optimizer=optimizer,
            loss='huber',
            metrics=['mae', 'mse']
        )

        self.model = model

        print(f"모델 구조 (24시간 예측 가능):")
        model.summary()

        return model

    def create_sequences_24h(self):
        """
        24시간 예측을 위한 시퀀스 생성 (기존 메서드 오버라이드)
        """
        print("=== 24시간 예측용 시퀀스 데이터 생성 ===")

        # 시간 기반 분할
        n = len(self.df)
        train_end = int(n * (1 - self.test_size - self.val_size))
        val_end = int(n * (1 - self.test_size))

        df_train = self.df.iloc[:train_end].copy()
        df_val = self.df.iloc[train_end:val_end].copy()
        df_test = self.df.iloc[val_end:].copy()

        print(f"데이터 분할: Train({len(df_train)}), Val({len(df_val)}), Test({len(df_test)})")

        # 특성 정규화
        train_features = df_train[self.feature_names].values
        self.feature_scaler.fit(train_features)

        # 타겟 정규화
        train_targets = df_train[self.target_col].values.reshape(-1, 1)
        self.target_scaler.fit(train_targets)

        # 24시간 예측용 시퀀스 생성
        self.X_train, self.y_train = self._build_sequences_24h(df_train)
        self.X_val, self.y_val = self._build_sequences_24h(df_val)
        self.X_test, self.y_test = self._build_sequences_24h(df_test)

        print(f"24시간 예측 시퀀스 생성 완료:")
        print(f"  Train: X{self.X_train.shape}, y{self.y_train.shape}")
        print(f"  Val:   X{self.X_val.shape}, y{self.y_val.shape}")
        print(f"  Test:  X{self.X_test.shape}, y{self.y_test.shape}")

    def _build_sequences_24h(self, df):
        """
        24시간 미래를 예측하는 시퀀스 생성
        """
        features = self.feature_scaler.transform(df[self.feature_names].values)
        targets = self.target_scaler.transform(df[self.target_col].values.reshape(-1, 1)).flatten()

        X, y = [], []
        for i in range(self.seq_len, len(df) - self.prediction_horizon + 1):
            # 과거 sequence_length 시간의 특성
            X.append(features[i-self.seq_len:i])
            # 미래 prediction_horizon 시간의 타겟
            y.append(targets[i:i+self.prediction_horizon])

        return np.array(X, dtype=np.float32), np.array(y, dtype=np.float32)

    def predict_24h_ahead(self, input_sequence):
        """
        24시간 미래 예측
        """
        # 입력 데이터 형태 확인 및 변환
        if len(input_sequence.shape) == 1:
            input_sequence = input_sequence.reshape(1, -1, 1)
        elif len(input_sequence.shape) == 2:
            input_sequence = input_sequence.reshape(1, input_sequence.shape[0], input_sequence.shape[1])

        # 예측 수행
        prediction_scaled = self.model.predict(input_sequence, verbose=0)

        # 역정규화
        prediction = self.target_scaler.inverse_transform(prediction_scaled.reshape(-1, 1))

        # 로그 변환 역변환
        if self.use_log_transform:
            prediction = np.expm1(prediction)

        return prediction.flatten()

    def detect_anomalies_threshold(self, actual, predicted, threshold_multiplier=2.5):
        """
        임계값 기반 이상 탐지
        """
        errors = np.abs(actual - predicted)

        if len(self.prediction_errors) > 0:
            threshold = np.mean(self.prediction_errors) + threshold_multiplier * np.std(self.prediction_errors)
        else:
            threshold = threshold_multiplier * np.std(errors)

        anomalies = errors > threshold
        return anomalies, errors, threshold

    def detect_anomalies_isolation_forest(self, data):
        """
        Isolation Forest를 사용한 이상 탐지
        """
        # 다차원 특성 구성
        features = []
        for i in range(len(data)):
            feature_vector = []

            # 현재 값
            feature_vector.append(data[i])

            # 최근 6시간 평균
            if i >= 6:
                feature_vector.append(np.mean(data[max(0, i-6):i+1]))
            else:
                feature_vector.append(data[i])

            # 변화율
            if i > 0:
                change_rate = (data[i] - data[i-1]) / (data[i-1] + 1e-8)  # 0으로 나누기 방지
                feature_vector.append(change_rate)
            else:
                feature_vector.append(0)

            # 최근 12시간 표준편차
            if i >= 12:
                feature_vector.append(np.std(data[max(0, i-12):i+1]))
            else:
                feature_vector.append(0)

            features.append(feature_vector)

        features = np.array(features)

        # Isolation Forest 학습 및 예측
        anomaly_labels = self.anomaly_detector.fit_predict(features)
        anomaly_scores = self.anomaly_detector.decision_function(features)

        return anomaly_labels == -1, anomaly_scores

    def add_context_features(self, data):
        """
        Context 정보를 데이터에 추가
        """
        data = data.copy()

        # hour_offset을 시간 정보로 변환
        if 'hour_offset' in data.columns:
            # 시간대별 컨텍스트
            data['hour_of_day'] = data['hour_offset'] % 24
            data['day_of_week'] = (data['hour_offset'] // 24) % 7
            data['is_weekend'] = (data['day_of_week'] >= 5).astype(int)

            # 피크/비피크 시간 구분
            data['time_context'] = 'normal'
            data.loc[data['hour_of_day'].isin([9, 10, 11, 14, 15, 16]), 'time_context'] = 'peak'
            data.loc[data['hour_of_day'].isin([0, 1, 2, 3, 4, 5, 22, 23]), 'time_context'] = 'low'
            data.loc[data['is_weekend'] == 1, 'time_context'] = 'weekend'

        return data

    def predict_with_context(self, input_sequence, context_type='normal'):
        """
        Context를 고려한 예측
        """
        base_prediction = self.predict_24h_ahead(input_sequence)

        # Context 가중치 적용
        context_weight = self.context_weights.get(context_type, 1.0)
        adjusted_prediction = base_prediction * context_weight

        return adjusted_prediction, base_prediction

    def update_prediction_errors(self, actual, predicted):
        """
        예측 오차 업데이트
        """
        if len(actual) != len(predicted):
            min_len = min(len(actual), len(predicted))
            actual = actual[:min_len]
            predicted = predicted[:min_len]

        errors = np.abs(actual - predicted)
        self.prediction_errors.extend(errors)

        # 최근 1000개 오차만 유지
        if len(self.prediction_errors) > 1000:
            self.prediction_errors = self.prediction_errors[-1000:]

    def evaluate_24h_model(self):
        """
        24시간 예측 모델 평가
        """
        print("=== 24시간 예측 모델 평가 ===")

        results = {}

        datasets = [
            ('Train', self.X_train, self.y_train),
            ('Validation', self.X_val, self.y_val),
            ('Test', self.X_test, self.y_test)
        ]

        for name, X, y_true in datasets:
            if X.shape[0] == 0:
                continue

            # 24시간 예측
            y_pred_scaled = self.model.predict(X, verbose=0)

            # 역정규화
            y_pred = self.target_scaler.inverse_transform(y_pred_scaled.reshape(-1, 1))
            y_actual = self.target_scaler.inverse_transform(y_true.reshape(-1, 1))

            # 로그 변환 역변환
            if self.use_log_transform:
                y_pred = np.expm1(y_pred)
                y_actual = np.expm1(y_actual)

            y_pred = y_pred.flatten()
            y_actual = y_actual.flatten()

            # 메트릭 계산
            mae = mean_absolute_error(y_actual, y_pred)
            rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
            r2 = r2_score(y_actual, y_pred)
            mape = np.mean(np.abs((y_actual - y_pred) / np.maximum(y_actual, 1))) * 100

            results[name.lower()] = {
                'mae': mae,
                'rmse': rmse,
                'r2': r2,
                'mape': mape,
                'y_true': y_actual,
                'y_pred': y_pred
            }

            print(f"{name:>10}: MAE={mae:.2f}, RMSE={rmse:.2f}, R²={r2:.4f}, MAPE={mape:.1f}%")

        return results

# MCP 알림 시스템 클래스 (Discord 버전)
class MCPAnomalyAlertSystem:
    """
    MCP 프로젝트용 이상 탐지 알림 시스템 - Discord Webhook
    """

    def __init__(self, alert_threshold=0.7, discord_webhook_url=None):
        self.alert_threshold = alert_threshold
        self.alert_history = []
        self.discord_webhook_url = discord_webhook_url

    def check_and_alert(self, anomaly_scores, current_values, predictions, timestamp=None):
        """
        이상 탐지 결과를 바탕으로 알림 생성
        """
        if timestamp is None:
            timestamp = datetime.now()

        alerts = []

        for i, (score, current, predicted) in enumerate(zip(anomaly_scores, current_values, predictions)):
            if score < -self.alert_threshold:  # Isolation Forest에서 음수가 이상
                severity = 'HIGH' if score < -1.0 else 'MEDIUM'

                alert = {
                    'timestamp': timestamp,
                    'severity': severity,
                    'anomaly_score': score,
                    'current_value': current,
                    'predicted_value': predicted,
                    'deviation': abs(current - predicted),
                    'hour_ahead': i + 1,
                    'message': f"시간 +{i+1}h: 예측값 {predicted:.1f}, 실제값 {current:.1f} (이상도: {score:.3f})"
                }
                alerts.append(alert)

        self.alert_history.extend(alerts)
        return alerts

    def format_discord_embed(self, alerts):
        """
        Discord Embed 형식 메시지 생성
        """
        if not alerts:
            return None

        high_alerts = [a for a in alerts if a['severity'] == 'HIGH']
        medium_alerts = [a for a in alerts if a['severity'] == 'MEDIUM']

        # Discord Embed 구조
        embed = {
            "title": "🚨 MCP 클라우드 리소스 이상 탐지 알림",
            "color": 15158332 if high_alerts else 16776960,  # 빨강(HIGH) 또는 노랑(MEDIUM)
            "fields": [],
            "footer": {
                "text": f"감지 시각: {alerts[0]['timestamp'].strftime('%Y-%m-%d %H:%M:%S')}"
            },
            "timestamp": alerts[0]['timestamp'].isoformat()
        }

        # HIGH 우선순위 알림
        if high_alerts:
            high_text = "\n".join([f"• {alert['message']}" for alert in high_alerts[:3]])
            embed["fields"].append({
                "name": "🔴 높은 우선순위 알림",
                "value": high_text,
                "inline": False
            })

        # MEDIUM 우선순위 알림
        if medium_alerts:
            medium_text = "\n".join([f"• {alert['message']}" for alert in medium_alerts[:3]])
            embed["fields"].append({
                "name": "🟡 중간 우선순위 알림",
                "value": medium_text,
                "inline": False
            })

        # 통계 정보
        embed["fields"].append({
            "name": "📊 탐지 통계",
            "value": f"총 {len(alerts)}개 이상 탐지 (HIGH: {len(high_alerts)}, MEDIUM: {len(medium_alerts)})",
            "inline": False
        })

        return {"embeds": [embed]}

    def send_discord_alert(self, alerts):
        """
        Discord Webhook으로 알림 전송
        """
        if not self.discord_webhook_url:
            print("Discord Webhook URL이 설정되지 않았습니다.")
            return False

        if not alerts:
            return False

        try:
            import requests

            payload = self.format_discord_embed(alerts)
            response = requests.post(self.discord_webhook_url, json=payload)

            if response.status_code == 204:
                print("Discord 알림 전송 성공")
                return True
            else:
                print(f"Discord 알림 전송 실패: {response.status_code}")
                return False

        except Exception as e:
            print(f"Discord 알림 전송 오류: {str(e)}")
            return False

# 실사용 예제 함수
def run_enhanced_mcp_predictor():
    """
    향상된 MCP 예측기 실행 예제
    """
    print("=== MCP 24시간 예측 및 이상 탐지 시스템 시작 ===")

    # 1. 모델 초기화
    predictor = EnhancedLSTMPredictor(
        sequence_length=24,
        target_col='total_events',
        prediction_horizon=24,  # 24시간 예측
        use_log_transform=True,
        handle_outliers=True
    )

    # 2. 데이터 로드 (기존 메서드 사용)
    print("\n데이터 로드 및 전처리...")
    df = predictor.load_and_prepare_data()

    # 3. Context 정보 추가
    df_with_context = predictor.add_context_features(df)
    predictor.df = df_with_context

    # 4. 24시간 예측용 시퀀스 생성
    print("\n24시간 예측용 시퀀스 생성...")
    predictor.create_sequences_24h()

    # 5. 24시간 예측 모델 구축
    print("\n24시간 예측 모델 구축...")
    predictor.build_model(
        lstm_units=[64, 32],
        dense_units=[32, 16],
        dropout_rate=0.3
    )

    # 6. 모델 훈련
    print("\n모델 훈련...")
    history = predictor.train_model(epochs=50, batch_size=32, patience=15)

    # 7. 모델 평가
    print("\n24시간 예측 모델 평가...")
    results = predictor.evaluate_24h_model()

    # 8. 이상 탐지 테스트
    print("\n이상 탐지 테스트...")
    if predictor.X_test.shape[0] > 0:
        # 테스트 데이터의 마지막 시퀀스로 24시간 예측
        test_sequence = predictor.X_test[-1:]  # 마지막 시퀀스 하나 (슬라이싱 수정)

        # 입력 형태 확인
        print(f"테스트 시퀀스 형태: {test_sequence.shape}")

        if test_sequence.shape[0] == 0:
            print("경고: 테스트 시퀀스가 비어있습니다.")
            return predictor, results

        prediction_24h = predictor.predict_24h_ahead(test_sequence)

        print(f"24시간 예측 결과: {prediction_24h[:6]} ... (첫 6시간)")

        # 실제 값과 비교 (가능한 경우)
        if len(predictor.y_test) > 0:
            actual_24h = predictor.target_scaler.inverse_transform(predictor.y_test[-1:].reshape(-1, 1))
            if predictor.use_log_transform:
                actual_24h = np.expm1(actual_24h)
            actual_24h = actual_24h.flatten()

            # 이상 탐지 수행
            anomalies_iso, scores_iso = predictor.detect_anomalies_isolation_forest(actual_24h)

            print(f"이상 탐지 결과: {np.sum(anomalies_iso)}개 이상 포인트 발견")

            # 알림 시스템 테스트
            alert_system = MCPAnomalyAlertSystem(
                discord_webhook_url="https://discord.com/api/webhooks/1423316756906115084/PdlOaM2SMh3u-46Ua0ftKEFBIUaDyovrscx3n-mP7-dclbV_iPioporRrAKFS7UT-zx4"
            )
            alerts = alert_system.check_and_alert(
                scores_iso, actual_24h, prediction_24h
            )

            if alerts:
                alert_system.send_discord_alert(alerts)
                print(alert_system.format_slack_message(alerts))
            else:
                print("\n정상 범위 내 동작 - 알림 없음")

    # 9. 결과 시각화
    print("\n결과 시각화...")
    predictor.plot_comprehensive_results(results)

    print("\n=== MCP 24시간 예측 시스템 완료 ===")
    return predictor, results

# 실행 코드를 기존 main() 함수 뒤에 추가
if __name__ == "__main__":
    # 기존 코드 실행
    try:
        predictor, results = main()
        print("\n" + "="*60)
        print("기존 모델 완료. 이제 24시간 예측 모델을 시작합니다...")
        print("="*60)

        # 새로운 24시간 예측 모델 실행
        enhanced_predictor, enhanced_results = run_enhanced_mcp_predictor()

    except Exception as e:
        print(f"오류 발생: {e}")
        # 기존 모델이 실패해도 새 모델은 실행
        print("기존 모델 실행 실패. 24시간 예측 모델만 실행합니다...")
        enhanced_predictor, enhanced_results = run_enhanced_mcp_predictor()
```

```python
"""
FastAPI 기반 LSTM 24시간 예측 및 이상 탐지 API 서버
기존 EnhancedLSTMPredictor 클래스를 사용하여 REST API 제공
"""

from fastapi import FastAPI, HTTPException, BackgroundTasks
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel, Field
from typing import List, Optional, Dict, Any
import numpy as np
import pickle
from datetime import datetime
import uvicorn
import logging

# 로깅 설정
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# FastAPI 앱 초기화
app = FastAPI(
    title="MCP LSTM 예측 API",
    description="24시간 클라우드 리소스 예측 및 이상 탐지 API",
    version="1.0.0"
)

# CORS 설정 (프론트엔드 연동용)
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # 실제 배포시에는 특정 도메인으로 제한
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# 전역 변수로 모델 저장
predictor = None
model_loaded = False

# ==================== Pydantic 모델 정의 ====================

class HealthResponse(BaseModel):
    status: str
    model_loaded: bool
    timestamp: str

class PredictionInput(BaseModel):
    """24시간 예측 입력"""
    sequence: List[List[float]] = Field(
        ...,
        description="과거 24시간의 특성 데이터 [timesteps, features]"
    )
    context: Optional[str] = Field(
        "normal",
        description="시간 컨텍스트: peak, normal, low, weekend"
    )

class PredictionOutput(BaseModel):
    """24시간 예측 출력"""
    predictions: List[float] = Field(..., description="24시간 예측값")
    base_predictions: Optional[List[float]] = Field(None, description="Context 적용 전 예측값")
    context_applied: str
    timestamp: str

class AnomalyDetectionInput(BaseModel):
    """이상 탐지 입력"""
    data: List[float] = Field(..., description="검사할 시계열 데이터")
    method: str = Field("both", description="탐지 방법: threshold, isolation_forest, both")

class AnomalyDetectionOutput(BaseModel):
    """이상 탐지 출력"""
    anomalies_threshold: Optional[List[bool]] = None
    anomalies_isolation: Optional[List[bool]] = None
    anomaly_scores: Optional[List[float]] = None
    threshold_value: Optional[float] = None
    anomaly_count: int
    timestamp: str

class BatchPredictionInput(BaseModel):
    """배치 예측 입력"""
    sequences: List[List[List[float]]] = Field(
        ...,
        description="여러 시퀀스 [[seq1], [seq2], ...]"
    )
    contexts: Optional[List[str]] = Field(None, description="각 시퀀스의 컨텍스트")

class ModelMetrics(BaseModel):
    """모델 성능 지표"""
    train_mae: float
    train_rmse: float
    train_r2: float
    test_mae: float
    test_rmse: float
    test_r2: float
    model_info: Dict[str, Any]

# ==================== 유틸리티 함수 ====================

def load_model_from_file(model_path: str = "complete_mcp_lstm.h5",
                        metadata_path: str = "mcp_model_metadata.pkl"):
    """저장된 모델과 메타데이터 로드"""
    global predictor, model_loaded

    try:
        import tensorflow as tf

        # 메타데이터 로드
        with open(metadata_path, 'rb') as f:
            metadata = pickle.load(f)

        # 예측기 인스턴스 생성
        from __main__ import EnhancedLSTMPredictor  # 기존 클래스 import

        predictor = EnhancedLSTMPredictor(
            sequence_length=metadata['sequence_length'],
            target_col=metadata['target_col'],
            prediction_horizon=24,
            use_log_transform=metadata['use_log_transform']
        )

        # 저장된 속성 복원
        predictor.feature_names = metadata['feature_names']
        predictor.feature_scaler = metadata['feature_scaler']
        predictor.target_scaler = metadata['target_scaler']

        # 모델 로드
        predictor.model = tf.keras.models.load_model(model_path)

        model_loaded = True
        logger.info(f"모델 로드 완료: {model_path}")
        return True

    except Exception as e:
        logger.error(f"모델 로드 실패: {str(e)}")
        model_loaded = False
        return False

def validate_input_shape(sequence: List[List[float]], expected_features: int) -> bool:
    """입력 데이터 형태 검증"""
    if len(sequence) != predictor.seq_len:
        return False
    if len(sequence[0]) != expected_features:
        return False
    return True

# ==================== API 엔드포인트 ====================

@app.on_event("startup")
async def startup_event():
    """서버 시작 시 모델 로드"""
    logger.info("MCP LSTM API 서버 시작 중...")
    load_model_from_file()

@app.get("/", response_model=HealthResponse)
async def root():
    """헬스체크 엔드포인트"""
    return {
        "status": "running",
        "model_loaded": model_loaded,
        "timestamp": datetime.now().isoformat()
    }

@app.get("/health", response_model=HealthResponse)
async def health_check():
    """상세 헬스체크"""
    return {
        "status": "healthy" if model_loaded else "model_not_loaded",
        "model_loaded": model_loaded,
        "timestamp": datetime.now().isoformat()
    }

@app.post("/predict", response_model=PredictionOutput)
async def predict_24h(input_data: PredictionInput):
    """
    24시간 예측 엔드포인트

    - **sequence**: 과거 24시간의 특성 데이터 (shape: [24, n_features])
    - **context**: 시간 컨텍스트 (peak/normal/low/weekend)
    """
    if not model_loaded:
        raise HTTPException(status_code=503, detail="모델이 로드되지 않았습니다")

    try:
        # 입력 데이터 검증
        sequence = np.array(input_data.sequence)
        expected_features = len(predictor.feature_names)

        if not validate_input_shape(input_data.sequence, expected_features):
            raise HTTPException(
                status_code=400,
                detail=f"입력 형태 오류: 예상 [{predictor.seq_len}, {expected_features}], "
                       f"실제 {sequence.shape}"
            )

        # Context 적용 예측
        adjusted_pred, base_pred = predictor.predict_with_context(
            sequence,
            input_data.context
        )

        return {
            "predictions": adjusted_pred.tolist(),
            "base_predictions": base_pred.tolist(),
            "context_applied": input_data.context,
            "timestamp": datetime.now().isoformat()
        }

    except Exception as e:
        logger.error(f"예측 오류: {str(e)}")
        raise HTTPException(status_code=500, detail=f"예측 실패: {str(e)}")

@app.post("/predict/batch", response_model=List[PredictionOutput])
async def predict_batch(input_data: BatchPredictionInput):
    """
    배치 예측 엔드포인트
    여러 시퀀스를 한 번에 예측
    """
    if not model_loaded:
        raise HTTPException(status_code=503, detail="모델이 로드되지 않았습니다")

    try:
        results = []
        contexts = input_data.contexts or ["normal"] * len(input_data.sequences)

        for seq, ctx in zip(input_data.sequences, contexts):
            sequence = np.array(seq)
            adjusted_pred, base_pred = predictor.predict_with_context(sequence, ctx)

            results.append({
                "predictions": adjusted_pred.tolist(),
                "base_predictions": base_pred.tolist(),
                "context_applied": ctx,
                "timestamp": datetime.now().isoformat()
            })

        return results

    except Exception as e:
        raise HTTPException(status_code=500, detail=f"배치 예측 실패: {str(e)}")

@app.post("/anomaly/detect", response_model=AnomalyDetectionOutput)
async def detect_anomalies(input_data: AnomalyDetectionInput):
    """
    이상 탐지 엔드포인트

    - **data**: 검사할 시계열 데이터
    - **method**: threshold (임계값), isolation_forest (ML), both (둘 다)
    """
    if not model_loaded:
        raise HTTPException(status_code=503, detail="모델이 로드되지 않았습니다")

    try:
        data = np.array(input_data.data)
        result = {
            "timestamp": datetime.now().isoformat()
        }

        if input_data.method in ["threshold", "both"]:
            # 임계값 기반 탐지 (예측값 필요)
            # 간단히 평균값을 예측으로 사용
            predicted = np.full_like(data, data.mean())
            anomalies, errors, threshold = predictor.detect_anomalies_threshold(
                data, predicted
            )
            result["anomalies_threshold"] = anomalies.tolist()
            result["threshold_value"] = float(threshold)

        if input_data.method in ["isolation_forest", "both"]:
            # Isolation Forest 기반 탐지
            anomalies_iso, scores_iso = predictor.detect_anomalies_isolation_forest(data)
            result["anomalies_isolation"] = anomalies_iso.tolist()
            result["anomaly_scores"] = scores_iso.tolist()

        # 이상 카운트
        if "anomalies_threshold" in result:
            result["anomaly_count"] = int(np.sum(result["anomalies_threshold"]))
        elif "anomalies_isolation" in result:
            result["anomaly_count"] = int(np.sum(result["anomalies_isolation"]))

        return result

    except Exception as e:
        raise HTTPException(status_code=500, detail=f"이상 탐지 실패: {str(e)}")

@app.get("/model/info")
async def get_model_info():
    """모델 정보 조회"""
    if not model_loaded:
        raise HTTPException(status_code=503, detail="모델이 로드되지 않았습니다")

    return {
        "model_type": "LSTM",
        "prediction_horizon": predictor.prediction_horizon,
        "sequence_length": predictor.seq_len,
        "n_features": len(predictor.feature_names),
        "feature_names": predictor.feature_names,
        "target_column": predictor.target_col,
        "use_log_transform": predictor.use_log_transform,
        "context_weights": predictor.context_weights
    }

@app.post("/model/reload")
async def reload_model(background_tasks: BackgroundTasks):
    """모델 재로드"""
    background_tasks.add_task(load_model_from_file)
    return {"message": "모델 재로드 시작"}

@app.get("/metrics")
async def get_metrics():
    """모델 성능 지표 조회"""
    if not model_loaded:
        raise HTTPException(status_code=503, detail="모델이 로드되지 않았습니다")

    # 실제 메트릭은 훈련 시 저장된 값을 사용
    return {
        "message": "메트릭 조회는 별도 파일에서 로드 필요",
        "model_info": {
            "sequence_length": predictor.seq_len,
            "prediction_horizon": predictor.prediction_horizon,
            "n_features": len(predictor.feature_names)
        }
    }

# ==================== 실행 코드 ====================

if __name__ == "__main__":
    # 개발 서버 실행
    uvicorn.run(
        "api_server:app",  # 파일명이 api_server.py인 경우
        host="0.0.0.0",
        port=8000,
        reload=True,  # 코드 변경 시 자동 재시작
        log_level="info"
    )
```
